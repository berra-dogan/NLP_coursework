```python
import pandas as pd
from ast import literal_eval

main = pd.read_csv(
        "data/dontpatronizeme_pcl.tsv",
        sep="\t",
        skiprows=4,
        engine="python",
        # index_col=0,
        header =None,
        names = ["par_id","art_id","keyword","country","text","label"]
    )
# main.set_index(False)
val = pd.read_csv("data/val_labels.csv")   # columns: par_id, label
```

```python
main.head()
```

```python
val.head()
```

```python
val["label_vec"] = val["label"].apply(literal_eval)

val_ids = set(val["par_id"].astype(int))
main["par_id"] = main["par_id"].astype(int)

val_df = main[main["par_id"].isin(val_ids)].copy()
train_df = main[~main["par_id"].isin(val_ids)].copy()

train_df.to_csv("data/PCL_train_dataset.tsv", sep="\t", index=False)
val_df.to_csv("data/PCL_val_dataset.tsv", sep="\t", index=False)

print("train:", len(train_df), "val:", len(val_df))

missing = val_ids - set(val_df["par_id"])
print("val ids missing from main file:", len(missing))
```
